```python
import pandas as pd
```

```python
UserTweets = pd.read_csv('/home/is/shuntaro-o/dev/disaster_analysis_Twitter/data/Foursquare_Osaka/FoursquareUserTweetsOsaka_20210511.txt', sep=',', lineterminator='\n', header=None, names=['local_id', 'tweet_id', 'timestamp', 'user_id', 'user_name', 'user_screenname', 'source', 'text', 'place_id', 'longitude', 'latitude'])
```

```python
UserPlase = pd.read_csv('/home/is/shuntaro-o/dev/disaster_analysis_Twitter/data/Foursquare_Osaka/FoursquareUserPlacesOsaka_20210511.txt', sep=',', lineterminator='\n', header=None,names = ['place_id', 'city_name', 'object', 'country', 'UserPlase_longitude', 'UserPlase_latitude', 'UserPlase_sub_longitude', 'UserPlase_sub_latitude'])
```

```python
UserPlase =UserPlase[~UserPlase.duplicated()]
```

```python
merged_Tweet_place = pd.merge(UserTweets , UserPlase, how='inner', on = "place_id")
```

```python
merged_Tweet_place = merged_Tweet_place.reset_index()
```
